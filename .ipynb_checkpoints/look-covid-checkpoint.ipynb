```python
import statistics
from datetime import datetime as dt

#df_temps.columns.values
do_not_include = ['Province/State', 'Country/Region', 'Lat', 'Long']

for x in df_temps.columns.values:
    if (x not in do_not_include):
        t = pd.to_datetime(df_temps[x].name)
        t = int(t.value / 10**9)
        #print(t, type(t))
        time_values = []
        for time_loop in range(14):
            the_stamp = t - (one_day * time_loop)
            time_values.append(str(the_stamp))
        #print(time_values)
        
        for y in range(df_temps['1/22/20'].values.size):
            temperature_data = []
            humidity_data = []
            latitude = str(df_temps.iloc[y][2])
            longitude = str(df_temps.iloc[y][3])
            #print (latitude, longitude)
            
            temperature_data = []
            humidity_data = []
            
            for z in time_values:
                url = 'https://api.darksky.net/forecast/' + API_KEY + '/' + latitude + "," + longitude + ',' + z
                url = url + '?exclude=currently,flags&units=si'
                #print(url, x, longitude, latitude)
                response = requests.get(url)
                info = json.loads(response.content)
                tMax = info['daily']['data'][0]['temperatureHigh']
                hum = info['daily']['data'][0]['humidity'] * 100
                temperature_data.append(tMax)
                humidity_data.append(hum)
            df_temps.at[y, x] = statistics.median(temperature_data)
            df_humid.at[y, x] = statistics.median(humidity_data)
            #print(df_temps.iloc[y][x])
            if (y == 4):
                break
    if (x == '1/22/20'):
        break

#print(temperature_data, humidity_data)
```
